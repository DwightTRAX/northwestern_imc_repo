### Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.20, random_state=11, stratify=y)
```

```python
# For evaluating the model
def scoring(y_test, pred):
    print('Accuracy: {}'.format(metrics.accuracy_score(y_test, pred)))
    print('roc_auc_score : ', metrics.roc_auc_score(y_test, pred))
```

### Machine Learning Models

### Logistic Regression

```python
# Baseline Logistic regression model
m = LogisticRegression()
m.fit(X_train, y_train)
pred = m.predict(X_test)
scoring( y_test, pred)
```

### Logistic Regression GridSearch & CrossValidation

```python
parameters = {'C':[0.5,1, 10, 20, 30]}
clf = GridSearchCV(LogisticRegression(class_weight = 'balanced', penalty = 'l2'), parameters, cv=5,scoring='roc_auc',verbose=1)
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

### Logistic Regression Evaluation

```python
m = LogisticRegression(C=0.5, class_weight = 'balanced', penalty ='l2')
m.fit(X_train, y_train)
pred = m.predict(X_test)
scoring( y_test, pred)
```

```python
## Important features
print([X_train.columns[i] for i in np.argsort(m.coef_[0])[::-1]])
```

```python
# Confusion matrix
cm = metrics.confusion_matrix(y_test, pred)
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".0f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(metrics.accuracy_score( y_test, pred))
plt.title(all_sample_title, size = 15);
```

```python
probs = clf.predict_proba(X_test)
preds = probs[:,1]
fpr_test, tpr_test, threshold = metrics.roc_curve(y_test, preds)
roc_auc_test = metrics.auc(fpr_test, tpr_test)

probs = clf.predict_proba(X_train)
preds = probs[:,1]
fpr_train, tpr_train, threshold = metrics.roc_curve(y_train, preds)
roc_auc_train = metrics.auc(fpr_train, tpr_train)

# method I: plt
import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr_test, tpr_test, 'b', label = 'test AUC = %0.3f' % roc_auc_test)
plt.plot(fpr_train, tpr_train, 'g', label = 'train AUC = %0.3f' % roc_auc_train)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
# Plot feature importance
feature_importance = m.feature_importances_
# make importances relative to max importance
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5
plt.subplot(1, 2, 2)
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, X_test.columns[sorted_idx])
plt.xlabel('Relative Importance')
plt.title('Variable Importance')
plt.show()
```

```python
sca_gb = GradientBoostingClassifier(n_estimators=1250, learning_rate=0.05,
max_features='sqrt', min_samples_leaf=1,min_samples_split = 2)
std_clf = make_pipeline(StandardScaler(), sca_gb)
std_clf.fit(X_train, y_train)
pred_test_std = std_clf.predict(X_test)
```

```python
scoring(y_test, pred_test_std)
cm = metrics.confusion_matrix(y_test, pred_test_std)
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".0f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(metrics.accuracy_score( y_test, pred))
plt.title(all_sample_title, size = 15);
```

```python
probs = std_clf.predict_proba(X_test)
preds = probs[:,1]
fpr_test, tpr_test, threshold = metrics.roc_curve(y_test, preds)
roc_auc_test = metrics.auc(fpr_test, tpr_test)

probs = m.predict_proba(X_train)
preds = probs[:,1]
fpr_train, tpr_train, threshold = metrics.roc_curve(y_train, preds)
roc_auc_train = metrics.auc(fpr_train, tpr_train)



# method I: plt
import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr_test, tpr_test, 'b', label = 'test AUC = %0.3f' % roc_auc_test)
plt.plot(fpr_train, tpr_train, 'g', label = 'train AUC = %0.3f' % roc_auc_test)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')//;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;//'
plt.xlabel('False Positive Rate')
plt.show()
```
